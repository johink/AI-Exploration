```python
from blackjack_rl import Blackjack, DealerRobot, HeuristicRobot, RLRobot, RandomChoiceRobot, Player
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
bj = Blackjack(2, 4, setup = False)

p1 = Player("Random", 1e6, True, RandomChoiceRobot())
p2 = Player("Dealer", 1e6, True, DealerRobot())
p3 = Player("Heuristic", 1e6, True, HeuristicRobot())
p4 = Player("RL", 1e6, True, RLRobot(training_epochs=1000000))

bj.players = [p1,p2,p3,p4]
```

```python
bj.all_AI = True
```

```python
stats = bj.play_n_games(10000)
stats = np.array(stats)
```

```python
fig, ax = plt.subplots()
for i in range(len(bj.players)):
    ax.plot(stats[:,i], label = bj.players[i].name)
ax.legend();
```
